```python
import numpy as np
import matplotlib.pyplot as plt #Imports necessary libraries

x=np.arange(-2*np.pi, 2*np.pi, 0.1) #Sets range of x axis
y=np.power(np.e,-x)*np.sin(x*2+(np.pi/3)) #Sets y to trig function

plt.plot(x,y) #Plots graph x to y
plt.show #Prints graph to user

#Graph Design
plt.title(r"y=$e^{-x}$ sin(2x+ $\frac{π}{3}$)")
plt.xlabel(r"-2$\pi$ ≤ x ≤ 2$\pi$")
```

```python
import random #Imports necessary libraries

digit1 = random.randint(0,9)
digit2 = random.randint(0,9)
digit3 = random.randint(0,9) #Creates 3 random numbers

digit1=str(digit1)
digit2=str(digit2)
digit3=str(digit3)  #Turns digits into strings so they can be concatenated

PIN = digit1+digit2+digit3 #Conatenate digits to form PIN 

print(PIN) #Prints PIN
```

```python
guessedPIN='000' #Sets Base for guessed PIN so guess can be incremented in while loop
while guessedPIN != PIN: #Creates loop until the guess equals the real PIN
    guessedPIN=int(guessedPIN) #Turns guessedPin to an Interger so it can be incremented
    guessedPIN=guessedPIN+1 #Increments guessedPin
    guessedPIN=str(guessedPIN) #Turns guessedPin back into a string so it can check if its the same as the real PIN
print(guessedPIN) #Prints guessed PIN once the guess is correct
```
